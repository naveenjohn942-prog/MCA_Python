```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
calo=pd.read_csv('nutrients_csvfile.csv')
calo.head(10)
```

```python
calo['Category'].unique()
dp = calo[calo['Category'] == 'Dairy products']
```

```python
dp["Calories"] = pd.to_numeric(dp["Calories"], errors="coerce")
dp["Calories"].plot(kind="hist", label="Calories", color='b', fontsize=10, density=True)

plt.title('Histogram of Grams in Dairy Products')
plt.xlabel('Calories')

plt.show()
```

```python
dp.plot(x="Calories", y="Protein", kind="scatter",label='Calories',color='r')
plt.title('Scatter Plot - Setosa')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

```python
dp["Calories"].plot(kind="line",label='Calories',color='black',linewidth='1.5',ls='-')

plt.axis([0,50,2,1000])
plt.title('Line Plot')
plt.xlabel('Calories')
plt.show()
```

```python
dp=dp.drop(columns="User_ID")
```

```python
#area plot
dp.plot.area(stacked=False)
plt.show()
dp.describe()
```

```python
#bargraph
calo['Calories'] = pd.to_numeric(calo['Calories'], errors='coerce')

Type=list(calo['Category'].unique())
count=list(calo['Category'].value_counts())
Avg=list(calo.groupby('Category')["Calories"].mean())
plt.bar(Type, Avg, color=['maroon', 'yellow', 'blue', 'green', 'purple', 'orange', 'red'],width=0.4)
plt.title('Bar Plot')

plt.xlabel('Calories')
plt.ylabel('AverageCalories')
plt.xticks(rotation=90)

plt.show()
```

```python
#pie chart
ethnic_group_counts = dp['Calories'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(ethnic_group_counts, labels=ethnic_group_counts.index, autopct='%1.2f%%', startangle=140)
plt.title('Calories')
plt.axis('equal') 
plt.show()
```

```python
#box plot
dp.plot.box()
plt.title("Box Plot")
plt.legend(labels = ['Calories','Grams','Carbs','Protein'])

plt.show()
```

```python
import seaborn as sns
sns.set(style="white")
sns.pairplot(dp, hue="Category")
plt.show()
```

```python
#violen graph
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Calories', data=dp, palette='muted')
plt.title('Violin Plot of Calories by Gender')
plt.xlabel('Gender')
plt.ylabel('Calories')
plt.show()
```
